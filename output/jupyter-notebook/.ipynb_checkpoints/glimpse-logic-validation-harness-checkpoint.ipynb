# Experiment: Glimpse Logic Validation Harness

Objective:
- Verify that the browser runtime, master YAML, and safe function registry stay in sync.
- Inspect which rules fired, which functions were called, and how that changed lens and view outputs.


```python
from __future__ import annotations

import json
import subprocess
from pathlib import Path

ROOT = Path('c:/Users/USER/CascadeProjects')
REPORT_PATH = ROOT / 'tmp/jupyter-notebook/validation-report.json'
BOOTSTRAP = ROOT / 'scripts/bootstrap_glimpse_logic.mjs'

ROOT, REPORT_PATH, BOOTSTRAP
```

## Bootstrap

- Validate `glimpse.master.yaml`.
- Check function registry integrity.
- Run the sample datasets through the live pipeline.
- Write a compact JSON report for notebook inspection.


```python
completed = subprocess.run(
    ['node', str(BOOTSTRAP), '--report-json', str(REPORT_PATH)],
    cwd=ROOT,
    check=True,
    capture_output=True,
    text=True,
)
print(completed.stdout)
```

```python
report = json.loads(REPORT_PATH.read_text(encoding='utf-8'))
report.keys()
```

## Registry Inventory

- Confirm that the named safe functions exposed in YAML are also available in the runtime.
- Spot missing or misconfigured functions before they affect the browser UI.


```python
inventory = report['registry']['functions']
[(item['name'], item['returns'], ', '.join(item['scope'])) for item in inventory]
```

## Sample Dataset Runs

- Each entry shows the primary lens, secondary lenses, top views, evidence count, and fired rules.
- Compare these outputs before and after a rule change.


```python
sample_runs = report['sampleRuns']
for run in sample_runs:
    print(run['dataset'])
    print('  primary:', run['primaryLens'])
    print('  secondary:', ', '.join(run['secondaryLenses']) or 'none')
    print('  top views:', ', '.join(run['topViews']))
    print('  evidence:', run['evidenceCount'])
    print('  fired rules:', ', '.join(run['firedRules'][:8]))
    print()
```

## Validation And Regression Surface

- Use this section after adding or editing a function-backed rule.
- The target is deterministic behavior: same YAML plus same dataset should produce stable lenses and view ranking.


```python
validation = report['validationReport']
validation
```

```python
comparison = {
    run['dataset']: {
        'primaryLens': run['primaryLens'],
        'topViews': tuple(run['topViews']),
        'firedRuleCount': len(run['firedRules']),
        'diagnostics': len(run['validation'].get('diagnostics', [])),
    }
    for run in sample_runs
}
comparison
```
